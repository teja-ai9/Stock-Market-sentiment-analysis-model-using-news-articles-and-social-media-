```python
#subsetting necessary columns from the original Dataframe

df_var= df[["Vader Compound Score","Textblob Score","EmoLex Score", 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise', 'Trust']]
```

```python
df_var
```

```python
#Importing the necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Scaling the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled=pd.DataFrame(scaler.fit_transform(df_var),columns=df_var.columns)
```

```python
df_scaled
```

```python
#import stats models
#Splitting the data into train and test
from statsmodels import api as sm
from sklearn.model_selection import train_test_split

X = df_scaled
X = sm.add_constant(X)
y = df["Close"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#OLS Model fitting for the train data

model = sm.OLS(y_train,X_train).fit()
```

```python
model.summary2()
```

```python
#Subsetting the features - Anger, Fear, Textblob.
df_train_final1=df_scaled[['Textblob Score','Anger','Fear']]
```

```python
#OLS Model on the Test data

import statsmodels.api as sm
X1= sm.add_constant(df_train_final1)
ols=sm.OLS(y,X1)
result=ols.fit()
result.summary2()
```

```python
#Predicting the Stock Price using the OLS Model on the Test data

y_pred = model.predict(X_test)
y_pred
```

```python
#Predicting the Stock Price using the OLS Model on the Training data

y_pred1 = model.predict(X_train)
y_pred1
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
#Defining the variables for performance metrics - OLS Model

e1 = metrics.mean_absolute_error(y_test,y_pred)
e2 = np.mean(np.abs((y_test- y_pred)/y_test))*100
e3 = metrics.mean_squared_error(y_test, y_pred)
e4 = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
#creating a list for the evaluation metrics 

errors=['MAE','MAPE', 'MSE', 'RMSE']
values=[e1,e2,e3,e4]
```

```python
error = pd.DataFrame({'Errors' :errors, 'Values': values})
error
```

Regularized regression Model 

```python
# importing the Ridge model 

from sklearn.linear_model import Ridge

## training the model using Ridge 

ridgeReg = Ridge(alpha=0.05, normalize=True)

ridgeReg.fit(X_train,y_train)

pred = ridgeReg.predict(X_test)
```

```python
#Defining the variables for performance metrics - Regularized Regression Model

from sklearn import metrics
from sklearn.metrics import r2_score

ri1 = metrics.mean_absolute_error(y_test,pred)
ri2 = np.mean(np.abs(y_test- pred)/y_test)*100
ri3 = metrics.mean_squared_error(y_test, pred)
ri4 = np.sqrt(metrics.mean_squared_error(y_test, pred))
```

```python
values1=[ri1,ri2,ri3,ri4]
error = pd.DataFrame({'Errors' :errors, 'LR Metrics': values, 'Ridge Metrics': values1})

error
```

LassoRegression Model 

```python
#importing and training the data using LassoRegression Model 

from sklearn.linear_model import Lasso

lassoReg = Lasso(alpha=0.3, normalize=True)

lassoReg.fit(X_train,y_train)

pred1 = lassoReg.predict(X_test)
```

```python
#Defining the variables for performance metrics - Lasso Regression Model

from sklearn import metrics
from sklearn.metrics import r2_score

la1 = metrics.mean_absolute_error(y_test,pred1)
la2 = np.mean(np.abs(y_test- pred1)/y_test)*100
la3 = metrics.mean_squared_error(y_test, pred1)
la4 = np.sqrt(metrics.mean_squared_error(y_test, pred1))
```

```python
values2=[la1,la2,la3,la4]
error = pd.DataFrame({'Errors' :errors, 'LR Metrics': values, 'Ridge Metrics': values1, 'Lasso Metrics': values2})

error
```
