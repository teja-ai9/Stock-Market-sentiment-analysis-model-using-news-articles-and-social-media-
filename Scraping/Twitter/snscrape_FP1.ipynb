```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

#List for appending tweets data
attributes_list = []

#TwitterSearchScraper function to scrape relevant tweets and appending them to the list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#Reliance OR #RelianceIndustries since:2022-12-31 until:2023-1-8').get_items()):
    if i>1000000:
        break
    attributes_list.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.content, 
                                 tweet.user.followersCount])
    
#Dataframe for the list
tweets_df = pd.DataFrame(attributes_list, columns=["User", "Date Created", "Number of Likes", "Source of Tweet",
                                                        "Tweet", "Follower Count"])
```
